**Задание 1**
Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:

car_id = 'A222BC96’
Результат: Номер A222BС валиден. Регион: 96

car_id = 'АБ22ВВ193’
Результат: Номер не валиден

```python
import re
```

```python
car_id = 'А222ВС777'
```

```python
def validation_car_id(car_id):

    '''
    функция, которая принимает на вход строку и проверяет является ли она
    валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры).
    Обратите внимание, что не все буквы кириллического алфавита используются
    в транспортных номерах. Если номер валиден, то функция должна возвращать
    отдельно номер и регион.
    '''
    # Согласно требованиям ГОСТ, на автомобильных номерных знаках в России используются 
    # только те буквы, которые присутствуют и в кириллице, и в латинице. 
    # Таких букв только 12: А, В, Е, К, М, Н, О, Р, С, Т, У, Х. Подобное ограничение 
    # регулируется «Венской конвенцией о дорожном движении» 1968 года.

    car_id_pattern = r'^[АВЕКМНОРСТУХ]\d{3}(?<!000)[АВЕКМНОРСТУХ]{2}\d{2,3}$'    
    res = re.findall(car_id_pattern, car_id)
    if len(res) != 0:
        return(f'Номер {car_id} валиден. Регион: {car_id[6:]}')
    return(f'Номер {car_id} не валиден')

validation_car_id(car_id)        
```

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:

some_string = ‘Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.’

Результат:
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
some_string = '''Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.'''
```

```python
def eraser_repeat(some_string):

    '''функция, которая будет удалять все последовательные повторы слов 
    из заданной строки при помощи регулярных выражений.
    '''
    str_pattern = r'(\w+)(\s+\1)+'
    res_str = re.sub(str_pattern, r'\1', some_string)

    return res_str     

eraser_repeat(some_string)    
```

some_words = 'Информационные технологии’

```python
some_words = 'Институт мирового господства и новой картографии'
```

```python
def acronym(some_words):

    '''функция, которая будет возвращать акроним по 
    переданной в нее строке со словами.
    '''
    words_pattern = r'\b\w'
    res_words = ''.join(re.findall(r'\b\w', some_words, flags=re.IGNORECASE)).upper()
    return res_words

print(acronym(some_words))
```

**Задание 4**
Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:

emails = [‘test@gmail.com’, ‘xyz@test.in’, ‘test@ya.ru’, ‘xyz@mail.ru’, ‘xyz@ya.ru’, ‘xyz@gmail.com’]

Результат:

gmail.com: 2
test.in: 1
ya.ru: 2
mail.ru: 1

```python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
```

```python
def counting_domain(emails):

    '''функция, которая будет принимать на вход список email-адресов и 
    выводить их распределение по доменным зонам.    '''

    gmail = 0
    test_in = 0
    ya_ru = 0
    mail_ru = 0

    for e_address in emails:
        gmail_pattern = re.search(r'\w+@gmail\.com', e_address)
        test_in_pattern = re.search(r'\w+@test\.in', e_address)
        ya_ru_pattern = re.search(r'\w+@ya\.ru', e_address)
        mail_ru_pattern = re.search(r'\w+@mail\.ru', e_address)

        if gmail_pattern: gmail += 1
        if test_in_pattern: test_in += 1
        if ya_ru_pattern: ya_ru += 1
        if mail_ru_pattern: mail_ru += 1

    return(f'gmail.com: {gmail}  test.in: {test_in}  ya.ru: {ya_ru} mail.ru: {mail_ru}')
counting_domain(emails)     
```

```python
some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'
```

```python
def counting_vowel_consolant_words(some_text):

    '''функция, которая будет подсчитывать сколько слов начинается на гласные, 
    а сколько на согласные буквы в тексте (текст может быть написан как с 
    использованием букв кириллицы, так и латиницы).
    '''

    some_text_list = some_text.split()
    vowel_list = []
    for el in some_text_list:
        vowel_word_pattern = r'^[AEIOUYaeiouyАЕЁИОУЭЮЯаеёиоуэюя]\w+'
        res_vowel_word = re.findall(vowel_word_pattern, el)
        if len(res_vowel_word) != 0: 
            vowel_list.append(res_vowel_word)    
    return(f'Слов на гласные буквы: {len(vowel_list)}, Слов на согласные буквы: {len(some_text_list) - len(vowel_list)}')

counting_vowel_consolant_words(some_text)          
```
